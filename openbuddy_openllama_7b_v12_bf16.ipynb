```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
#OpenBuddy/openbuddy-openllama-7b-v12-bf16 -ответ на китайском
#openbuddy-atom-13b-v9-bf16
```

```python
tokenizer = AutoTokenizer.from_pretrained("OpenBuddy/openbuddy-mistral-7b-v17.1-32k", use_fast=False)
model = AutoModelForCausalLM.from_pretrained("OpenBuddy/openbuddy-mistral-7b-v17.1-32k", device_map="auto", torch_dtype="auto")
```

```python
request = "Напиши мне детальный план модели угроз безопасности информации объектов критической информационной инфраструктуры."

encoded_input = tokenizer(request, return_tensors='pt', add_special_tokens=False).to('cuda:0')
output = model.generate(
    **encoded_input,
    num_beams=2,
    do_sample=True,
    max_new_tokens=100
)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```
